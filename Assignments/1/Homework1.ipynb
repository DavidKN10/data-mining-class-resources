### Exercise 1: Show that the mean of the centered data matrix $\overline{\textbf{D}}$ in Eq.(1.9) is 0

$$\overline{\textbf{D}}=\textbf{D}-\textbf{1}\cdot \mu^T =
\begin{pmatrix}
 x_{1}^T\\  
 x_{2}^T\\
 \vdots\\ 
 x_{n}^T\\
\end{pmatrix}
-
\begin{pmatrix}
\mu ^T\\
\mu ^T\\
\vdots\\
\mu ^T\\
\end{pmatrix}
=
\begin{pmatrix}
 x_{1}^T - \mu ^T\\  
 x_{2}^T - \mu ^T\\
 \vdots\\ 
 x_{n}^T - \mu ^T\\
\end{pmatrix}
=
\begin{pmatrix}
 \overline{x}_1^T\\  
 \overline{x}_2^T\\
 \vdots\\ 
 \overline{x}_n^T\\
\end{pmatrix}
$$

<br>

You first start off by finding $mean(\overline{\textbf{D}})$
$$mean(\overline{\textbf{D}})=\frac{(x_1^T - \mu_1^T) + (x_2^T - \mu_2^T) + ... + (x_n^T - \mu_n^T)}{n}$$
Then add together all $x^T$ and all $\mu^T$
$$=\frac{(x_1^T + x_2^T+...+x_n^T) - (\mu^T + \mu^T +...+ \mu^T) }{n}$$

As you can notice, adding $\mu^T$ $n$ times will just equal $n\mu$
$$=\frac{(x_1^T + x_2^T+...+x_n^T) - n\mu }{n}$$

We can also find $n\mu$ by finding $mean(D)$ and solving for $n\mu$
$$mean(D)=\mu=\frac{x_1^T + x_2^T + ... + x_n^T}{n}$$
$$n\mu = x_1^T +x_2^T + ... +x_n^T$$

Since $n\mu = x_1^T +x_2^T + ... +x_n^T$, we can replace $(x_1^T + x_2^T+...+x_n^T)$ with $n\mu$. Now we can solve and see our result is 0.
$$\frac{(x_1^T + x_2^T+...+x_n^T) - n\mu }{n}=\frac{n\mu-n\mu}{n}$$
$$=\frac{0}{n}=0$$

### Exercise 12: Under what conditions will the covariance matrix $\textbf{\Sigma}$ be identical to the correlation matrix, whose $(i, j)$ entry gives the correlation between attributes $X_i$ and $X_j$? What can you conclude about the two variables?

<br>

Covariance matrix:
$$Cov(X_i, X_j) = \mathbb{E}((X_i-\mu_i)(X_j-\mu_j))$$
where $\mu=\mathbb{E}(x)$.

<br>

Correlation:
$$\rho_{ij}=\frac{Cov(X_i, X_j)}{\sigma_i \sigma_j}=\frac{\sigma_i \sigma_j}{\sqrt{\sigma_{ii} \sigma_{jj}}}$$
where $\sigma_i=\sqrt{Var(X_i)}=\sqrt{\sigma_{ii}}$.
<br>

The covariance matrix $\textbf{\Sigma}$ will be identical to the correlation matrix if all variables $X_i$ have the same variance and are correlated with each other. If all variables have the same variance, then $\sigma_{ii}=\sigma^2$ for all $i$. This simplifies the equation to:
$$\rho_{ij}=\frac{\sigma_{ij}}{\sigma^2}$$
If there was perfect correlation, the off-diagonal elements of the covariance matrix must equal the variance, implying that $\sigma_{ij}=\sigma^2$ for all $i\neq j$.

<br>

Therefore, we can conclude for both variables that they have the same variance $\sigma^2$ and that they are perfectly correlated either positively or negatively.

$$1-\frac{\sum ab}{\sqrt{\sum a^2}\sqrt{\sum b^2}}$$

#### Exercise 2: Show that the asymptotic volume of the hypersphere $\textbf{S}_d(r)$ for any value of radius $r$ eventually tends to zero as $d$ increases.

<br>

The volume of a hypersphere is given by:
$$V_d(r)=\frac{\pi^{d/2}}{\Gamma(\frac{d}{2}+1)}r^d$$

<br>

The gamma function $\Gamma(z)$ for very large $z$, can be approximated using Stirling's approximation:
$$\Gamma(z) \approx \sqrt{2\pi}z^{z-\frac{1}{2}}e^{-z}$$
For $z=\frac{d}{2}+1$, we get the following:
$$\Gamma(\frac{d}{2}+1) \approx \sqrt{2\pi}(\frac{d}{2}+1)^{\frac{d}{2}+\frac{1}{2}}e^{-(\frac{d}{2}+1)}$$
We can substitute the approximation into the volume formula:
$$V_d(r)=\frac{\pi^{d/2}}{\Gamma(\frac{d}{2}+1)}r^d \approx \frac{\pi^{d/2} r^d}{\sqrt{2\pi}(\frac{d}{2}+1)^{\frac{d}{2}+\frac{1}{2}}e^{-(\frac{d}{2}+1)}}$$
$$\approx \frac{\pi^{d/2} r^d e^{(\frac{d}{2}+1)}}{\sqrt{2\pi}(\frac{d}{2}+1)^{\frac{d}{2}+\frac{1}{2}}}$$
Comparing the numerator and denominator for very large $d$, we can conclude that the denominator grows at a faster rate because gamma functions grow faster than exponentials. When the denominator grows at a faster rate, the function will approach 0. 
<br>
Therefore, we can conclude that the asymptotic volume of the hypersphere  $\textbf{S}_d(r)$ tends to zero as the dimension increases.

#### a.) Compute the mean $\textbf{\mu}$ and covariance matrix $\textbf{\Sigma}$ for $\textbf{D}$.


```python
import numpy as np

#making D
D = np.array([
    [8, -20],
    [0, -1],
    [10, -19],
    [10, -20],
    [2, 0]
])

#finding mean
mu = np.mean(D, axis=0)

#finding covariance matrix
sigma = np.cov(D, rowvar=False)

#printing results
print("Mean:")
print(mu)
print()
print("Covariance Matrix:")
print(sigma)
```


#### b.) Compute the eigenvalues of $\Sigma$


```python
#computing eigenvalues
eigenvalues = np.linalg.eigvals(sigma)

#print results
print("Eigenvalues:")
print(eigenvalues)
```

#### c.) What is the "intrinsic" dimensionality of this dataset (discounting some small amount of variance)?
Our eigenvalues are $\lambda_1 = 1.332$, $\lambda_2 = 131.168$. Out total variance would be $\lambda_1+\lambda_2=1.332+131.168=132.5$. $\lambda_2$ makes up about 99% of the total variance. This suggests that the intrinsic dimensionality of the dataset is 1, discounting the variance captured by $\lambda_1$. 

<br>

#### d.) Compute the first principal component.


```python
#setting up eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(sigma)

#get largest index of eigenvalues
index_largest_eigenvalue = np.argmax(eigenvalues)

#calculate the first principal component
fpc = eigenvectors[:, index_largest_eigenvalue]

#printing results
print("First Principal Component:")
print(fpc)
```


#### e.) If the $\textbf{\mu}$ and $\textbf{\Sigma}$ from above characterize the normal distribution from which the points were generated, sketch the orientation/extent of the 2-dimensional normal density function.


```python
from matplotlib.patches import Ellipse
def plot_confidence_ellipse(mean, cov, ax, n_std=2.0, **kwargs):
    """
    Create a plot of the covariance confidence ellipse.
    Parameters:
    - mean : array-like, shape (2,)
    - cov : array-like, shape (2, 2)
    - ax : matplotlib axes object
    - n_std : float, number of standard deviations to determine the ellipse's radius
    - **kwargs : additional parameters to matplotlib.patches.Ellipse
    """
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0),
                      width=ell_radius_x * 2,
                      height=ell_radius_y * 2,
                      **kwargs)

    scale_x = np.sqrt(cov[0, 0]) * n_std
    scale_y = np.sqrt(cov[1, 1]) * n_std

    transf = plt.matplotlib.transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean[0], mean[1])

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)

# Plot the confidence ellipse
fig, ax = plt.subplots()
plot_confidence_ellipse(mu, sigma, ax, edgecolor='green')

# Re-plot data points and mean point
ax.scatter(D[:, 0], D[:, 1], color='blue', label='Data Points')
ax.scatter(mu[0], mu[1], color='red', label='Mean')

# Add labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data Points and Confidence Ellipse')
plt.legend()
plt.grid(True)
plt.axis('equal')

# Show the plot
plt.show()
```

Is there a different method of imputing values that would match the distribution more accurately? Describe your method. 

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

#column names of the dataset
column_names = ["displacement", "mpg", "cylinders", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]

#load data of the file
auto_mpg_dataset = pd.read_csv("auto-mpg\\auto-mpg.data", delim_whitespace=True, names=column_names, na_values='?')
#print(auto_mpg_dataset)

#replacing the missing values in the 'horsepower' column
auto_mpg_dataset["horsepower"] = pd.to_numeric(auto_mpg_dataset["horsepower"], errors="coerce")

print(auto_mpg_dataset)

#summary stats
summary_stats = auto_mpg_dataset.describe()
print(summary_stats)

#mean imputation
mean_imputer = SimpleImputer(strategy="mean")
auto_mpg_mean_imp = auto_mpg_dataset.copy()
auto_mpg_mean_imp["horsepower"] = mean_imputer.fit_transform(auto_mpg_mean_imp[["horsepower"]])

#median imputation
median_imputer = SimpleImputer(strategy="median")
auto_mpg_median_imp = auto_mpg_dataset.copy()
auto_mpg_median_imp["horsepower"] = median_imputer.fit_transform(auto_mpg_median_imp[["horsepower"]])

#mode imputation
mode_imputer = SimpleImputer(strategy="most_frequent")
auto_mpg_mode_imp = auto_mpg_dataset.copy()
auto_mpg_mode_imp["horsepower"] = mode_imputer.fit_transform(auto_mpg_mode_imp[["horsepower"]])

#calculating variance for imputed values
variance_original = auto_mpg_dataset["horsepower"].var()
variance_mean_imp = auto_mpg_mean_imp["horsepower"].var()
variance_median_imp = auto_mpg_median_imp["horsepower"].var() 
variance_mode_imp = auto_mpg_mode_imp["horsepower"].var()

#printing varainces
print("Original Variance:")
print(variance_original)
print()
print("Variance after mean imputation: ")
print(variance_mean_imp)
print()
print("Variance after median imputation: ")
print(variance_median_imp)
print()
print("Variance after mode imputation: ")
print(variance_mode_imp)
```

The imputation that results in the lowest variance is the mean imputation with a variance of 1459.18. Mean imputation tends to pull all values closer to the central mean, thereby reducing the overall spread (variance) of the dataset. This reduction occurs because the mean imputation uses the average of the observed values, which is less sensitive to extreme values than other measures like mode.


## 2.3 Problem 3

#### Load the iris sample dataset into Python using a Pandas dataframe. Perform a PCA using the Scikit Decomposition component, and provide the percentage of variance explained by each of the Principal Components. Compare this to the percentage of variance explained by each of the original features. What do you observe?

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA


#load dataset
iris = load_iris()
dataframe = pd.DataFrame(data=iris.data, columns=iris.feature_names)

#perform PCA
pca = PCA()
principal_components = pca.fit_transform(dataframe)

#calculate the percent of variance by each principal component
variance_explained = pca.explained_variance_ratio_

#calculate the percentage of variance explained by each original feature
variance_explained_original = dataframe.var() /dataframe.var().sum()

#create dataframes for the results
df_pca = pd.DataFrame({
    'Principal Component' : [f'PC{i+1}' for i in range(len(variance_explained))],
    'Variance Explained' : variance_explained
})

df_variance = pd.DataFrame({
    'Feature': dataframe.columns,
    'Variance Explained': variance_explained_original
})

#print dataframes
print("PCA Analysis:\n", df_pca)
print("\nOriginal Features Variance Explained:\n", df_variance)
```

The first principal component captures the vast majority of the variance in the dataset (~92%). The third and fourth principal components contribute very little to the overall variance. This indicates that PCA is highly effective in summarizing the data with fewer dimensions while retaining most of the information. The petal length explains the most variance among the original features. Sepal width explains the least variance among the original features. With this information, we can observe that the greatest cause for variance is petal length. This means that petal length changes the most when measuring different iris flowers.

## 2.5 Problem 5

#### Calculate the total variance of the original features and the total variance of the four eigenvectors from the above problem. What can you say about the corresponding values? If we wished to capture > 95% of the variance of the original data, how many principal components would we be selecting? How does this number correspond to the number of dimensions we are reducing our features to?

```python
#calculate the total variance of the original features
total_variance_original = dataframe.var().sum()

#calcualte the total variance of the four eigenvectors (principal components)
total_variance_pca = np.sum(pca.explained_variance_)

#determine the number of principal components required to capture > 95% of the variance
cumulative_variance_explained = np.cumsum(pca.explained_variance_ratio_)
num_components_95_variance = np.where(cumulative_variance_explained >= 0.95)[0][0] + 1

print("Total Variance of the original feature:", total_variance_original,"\n")
print("Total Variance explained by the principal components:",total_variance_pca,"\n")
print("Cumulative Variance explained by principal components (PC1, PC2, PC3, PC4):")
print(cumulative_variance_explained,"\n")
print("Number of principal components required to capture > 95% variance:", num_components_95_variance)
```

The total variance remains the same before and after the PCA transformation. This confirms that PCA is a linear transformation that retains the overall variance of the data. To capture more than 95% of the variance, we only need the first two principal components. This means we need can reduce the dimensionality of the dataset from four original features to two principal components while still retaining most of the information in the data.
